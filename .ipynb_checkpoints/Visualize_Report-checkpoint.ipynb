# Visualize Dakota report data from MongoDB

```python
%matplotlib inline
```

```python
#import dakota_report library
import dakota_report as dr
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load reports based on key values
docs = dr.find_documents({'group':'Genetic_param_study02','method':'vector_parameter_study'})
```

The following is the results of a 1D parameter study on each parameter about an approximated GIPL objective function minimum.

```python
for doc in docs:
    #Grab out data from documents
    out_dat = doc['out_data']
    obj_fn_dat = out_dat['obj_fn']
    for out in out_dat.keys():
        if (out != 'obj_fn'):
            
            print out
            plt.figure()
            plt.plot(out_dat[out],obj_fn_dat)
            plt.xlabel(out)
            #plt.ylabel(obj_fn_dat)
            plt.show()
            #print stats
            print 'Pre-objective: ' + str(doc['obj_fun_pre'])
            print 'Pre-params: ' + str(doc['pre_params'])
            print 'Post-params: '+ str(doc['post_params'])
            print 'Pre-error: ' + str(doc['pre_error'])
            print 'Post-error: ' + str(doc['post_error'])
            
```

The following is a parameter study to investigate the GIPL parameter equillibrium near the synthetic truth set. Each graph is the result of a parameter study on each of the 24 parameters in the GIPL model. The blue line is with all other parameters set to the synthetic values and thus we can observe a zero in the objective function. The green line is a parameter study starting at a random point within 7% tolerance of the synthetic truth values.

```python
#
off_on = dr.find_documents({'group':{'$in':['equi_param_study_OFFSET','equi_param_study_OFFSET_SYNTH']}})
```

```python
param_study_out = {'off':{},
                    'on':{}}
on_seed = {}
off_seed = {}
#Grab data from documents
for doc in off_on:
    #Grab out data from documents
    out_dat = doc['out_data']
    for out in out_dat.keys():
        #print out
        if (out != 'obj_fn'):
            if (doc['group']=='equi_param_study_OFFSET'):
                param_study_out['off'][out] = {'dat':[],
                                                'obj':[]}
                on_seed = doc['seed_pt']
                param_study_out['off'][out]['dat'] = out_dat[out]
                param_study_out['off'][out]['obj'] = out_dat['obj_fn'] 
            elif (doc['group']=='equi_param_study_OFFSET_SYNTH'):
                param_study_out['on'][out] = {'dat':[],
                                                'obj':[]}
                off_seed = doc['seed_pt']
                param_study_out['on'][out]['dat'] = out_dat[out]
                param_study_out['on'][out]['obj'] = out_dat['obj_fn']

print 'Start Graphing...'
#Graph data from parameter studies                
for key in param_study_out['on'].keys(): 
    print key
    plt.figure()
    #Normalize objective fn vals
    on_obj = param_study_out['on'][key]['obj']
    on_obj_norm = np.interp(on_obj,[np.amin(on_obj),np.amax(on_obj)],[0,1])
    off_obj = param_study_out['off'][key]['obj']
    off_obj_norm = np.interp(off_obj,[np.amin(off_obj),np.amax(off_obj)],[0,1])
    #Plot synth param study
    on_equi = plt.plot(param_study_out['on'][key]['dat'],on_obj_norm,label = 'Synth')
    off_equi = plt.plot(param_study_out['off'][key]['dat'],off_obj_norm,label = 'Offset')
    plt.legend(loc='upper_left')
    plt.xlabel(key)
    #plt.ylabel(obj_fn_dat)
    plt.show()
    print 'Synth value ' + key + ': ' + str(on_seed[key])
    print 'Offset value '+ key + ': ' + str(off_seed[key])
    #print stats
    #print 'Pre-objective: ' + str(doc['obj_fun_pre'])
    #print 'Pre-params: ' + str(doc['pre_params'])
    #print 'Post-params: '+ str(doc['post_params'])
    #print 'Pre-error: ' + str(doc['pre_error'])
    #print 'Post-error: ' + str(doc['post_error'])
```

The following is a parameter study in the scope of a 10% expansion about the synthetic parameter set. All parameters are incremented simultaneously essentially making a '45 degree' angle vector through the multidimensional parameter space. Hopefully we will be able to observe the effects of global error on the objective function behavior.

```python
param_study_all = dr.find_documents({'group':'synth_converge_all','method':'vector_parameter_study'})
```

```python
for doc in param_study_all:
    #Get out data
    out_dat = doc['out_data']
    #Get obj fun data
    obj_fun = out_dat['obj_fn']
    #Get one of the param's data
   
    param_dat = out_dat['kt1']
    #Normalize objective function
    obj_fun_norm = np.interp(obj_fun,[np.amin(obj_fun),np.amax(obj_fun)],[0,1])
    #Normalize parameter bounds to % offset
    param_dat_norm = np.interp(param_dat,[np.amin(param_dat),np.amax(param_dat)],[-doc['per_expan'],doc['per_expan']])
    
    #Graph results
    plt.plot(param_dat_norm,obj_fun_norm)
    plt.xlabel('Percent Offset')
    plt.ylabel('Objective Function')
```
